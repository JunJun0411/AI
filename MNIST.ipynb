```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from dataset.mnist import load_mnist
from PIL import Image
from knn import KNN
from tqdm import tqdm, trange , tqdm_notebook
    
(X_train, t_train), (X_test, t_test) = load_mnist(flatten=True, normalize=True)

y_name = np.unique(t_train) # y_name = [0, ... , 9]
knn_iris = KNN(3, X_train, t_train, y_name)


print("k = ", 3, ", weighted_majority_vote")
acc = 0
for i in tqdm_notebook(range(t_test.shape[0])):
    knn_iris.get_nearest_k(X_test[i])
    CC = knn_iris.weighted_majority_vote()
    TC = y_name[t_test[i]]
    if CC == TC:
        acc += 1 
    print("Test Data: ", i, " Computed class: ", CC, ",\tTrue class: ", TC)
    knn_iris.reset()
print("")
    
TESTCASE = 10000
print("weighted_majority_vote: ", acc/TESTCASE*100,"%")
```
