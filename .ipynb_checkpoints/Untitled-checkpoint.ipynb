```python
#1.6.1 단순한 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt

#데이터 준비
x = np.arange(0,6,0.1) # 0에서 6까지 0.1간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--" ,label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    sum = w1*x1 + w2*x2 - theta
    if sum >= 0:
        return 1
    else :
        return 0
    
print(AND(0,0))    
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
import numpy as np
def AND(x):
    w = np.array([0.6, 0.6])
    b = -1.0
    return (int(np.sum(w*x) + b > 0))
       
    
def OR(x):
    w = np.array([0.6, 0.6])
    b = -0.5
    return (int(np.sum(w*x) + b > 0))
    

def NAND(x):
    w = np.array([-0.6, -0.6])
    b = 1.0
    return (int(np.sum(w*x) + b > 0))

def XOR(x):#짜보기
    s1 = NAND(x)
    s2 = AND(x)
    return OR([s1, s2])
X = np.array([[0,0],
             [0,1],
             [1,0],
             [1,1]])
for i in range(X.shape[0]):
    print(AND(i), "")
for i in range(X.shape[0]):
    print(OR(i))
for i in range(X.shape[0]):
    print(NAND(i))
for i in range(X.shape[0]):
    print(XOR(i))
```

```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from dataset.mnist import load_mnist
from PIL import Image
from knn import KNN
from tqdm import tqdm, trange , tqdm_notebook

def img_show(img):
    pil_img = Image.fromarray(np.uint8(img))
    pil_img.show()

def KNN_Classification(K, X_train, y_train, y_name, X_test, y_test, sample, err):
    knn_iris = KNN(K, X_train, y_train, y_name)
    knn_iris.show_dim()
    print("k = ", K, ", majority_vote")
    err1=0
    for i, sa in enumerate(sample):
        knn_iris.get_nearest_k(X_test[i])
        CC = knn_iris.majority_vote()
        TC = y_name[y_test[i]]
        if CC == TC:
            err1 += 1 
        print(sa, " th data result: ", CC, ",\tlable: ", TC)
        knn_iris.reset()
    err.append(err1)
    err1 = 0
    print("k = ", K, ", weighted_majority_vote")
    for i, sa in enumerate(sample):
        knn_iris.get_nearest_k(X_test[i])
        CC = knn_iris.weighted_majority_vote()
        TC = y_name[y_test[i]]
        if CC == TC:
            err1 += 1 
        print(sa, " th data result: ", CC, ",\tlable: ", TC)
        knn_iris.reset()
    err.append(err1)
    print("")

def Hand(X_test):
    #Hand_crafted feature t_test
    qwe = X_test[0].reshape(28,-1)
    t_test = np.copy(np.array([np.mean(qwe[i-2:i, j-2:j])\
                                         for i in range(2, 29, 2) for j in range(2, 29, 2)]))
    for i in range(1, X_test.shape[0]):
        qwe = X_test[i].reshape(28,-1)
        t_test = np.r_[t_test, np.array([np.mean(qwe[i-2:i, j-2:j])\
                                         for i in range(2, 29, 2) for j in range(2, 29, 2)])]
    t_test = t_test.reshape(-1,196)
    return t_test
    
(x_train, t_train), (X_test, y_test) = load_mnist(flatten=False, normalize=True)

image = Hand(x_train)
label = t_train
y_name = np.unique(label)
print(image.shape)


size = 100
sample = np.random.randint(0, X_test.shape[0], size)
t_test = Hand(X_test[sample])
print(t_test.shape)
print(y_test[sample])
err=[]

KNN_Classification(3, image, label, y_name, t_test, y_test[sample], sample, err)    # K=3일 때
print("Majority_Vote accuracy \t", err[0]/size*100,"%")
print("Weighted_Majority_Vote accuracy", err[1]/size*100,"%")
```
