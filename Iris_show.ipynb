```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from knn import KNN

iris = load_iris()
# print(iris)

X = iris.data #[:, :2]  # iris data input

# y의 label을 두개로 하고싶다면
# y = np.where(iris.target==2, 0, iris.target) 
y = iris.target # iris target (label)

y_name = iris.target_names # label 2개로 할 땐 [:2] 붙여주자
# print(y_name)

l = 15
for_test = np.array([i%l == (l-1) for i in range(y.shape[0])])
for_train = ~for_test
# print(for_test)

X_train = X[for_train]
y_train = y[for_train]
# print(X.shape)
# print(X_train.shape)

X_test = X[for_test]
y_test = y[for_test]
# print(X_test)
# print(y_test)

knn_iris = KNN(3, X_train, y_train, y_name)
knn_iris.show_dim()
print("k = 3, majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()
print("k = 3, weighted_majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.weighted_majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()
print("")

knn_iris = KNN(5, X_train, y_train, y_name)
knn_iris.show_dim()
print("k = 5, majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()

print("k = 5, weighted_majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.weighted_majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()
    
print("")
knn_iris = KNN(10, X_train, y_train, y_name)
knn_iris.show_dim()
print("k = 10, majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()

print("k = 10, weighted_majority_vote")
for i in range(y_test.shape[0]):
    knn_iris.get_nearest_k(X_test[i])
    print("Test Data: ", i, " Computed class: ", knn_iris.weighted_majority_vote(),
         ",\tTrue class: ", y_name[y_test[i]])
    knn_iris.reset()

# x1_min, x1_max = X[:, 0].min() - .5, X[:, 0].max() + .5
# x2_min, x2_max = X[:, 1].min() - .5, X[:, 1].max() + .5
# plt.figure(2, figsize=(8, 6))
# # Plot the training points
# plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
#  edgecolor='k')
# plt.xlabel('Sepal length')
# plt.ylabel('Sepal width')
# plt.xlim(x1_min, x1_max)
# plt.ylim(x2_min, x2_max)
# plt.xticks(())
# plt.yticks(())
# plt.show()
```
